Exploratory Data Analysis (EDA) adalah proses analisis data untuk memahami struktur, pola, dan hubungan dalam dataset. Berikut adalah beberapa langkah umum dalam melakukan EDA:

1. **Pemahaman Awal Dataset**:
   - Baca dokumentasi dataset (jika tersedia) untuk memahami arti dari setiap kolom.
   - Tampilkan beberapa baris pertama dari dataframe untuk melihat struktur data.

2. **Pemeriksaan Tipe Data**:
   - Periksa tipe data dari setiap kolom (numerik, kategorikal, dll.).
   - Pastikan tipe data telah diinterpretasikan dengan benar oleh pandas.

3. **Penanganan Missing Values**:
   - Identifikasi kolom yang memiliki nilai yang hilang (NaN/null).
   - Tentukan strategi penanganan nilai yang hilang, seperti menghapus baris atau mengisi dengan nilai tertentu.

4. **Statistik Deskriptif**:
   - Hitung statistik deskriptif untuk variabel numerik (mean, median, standar deviasi, dll.).
   - Hitung distribusi variabel kategorikal (jumlah kategori, frekuensi masing-masing kategori, dll.).

5. **Visualisasi Data**:
   - Gunakan plot seperti histogram, box plot, dan scatter plot untuk memvisualisasikan distribusi dan hubungan antar variabel.
   - Gunakan teknik visualisasi yang sesuai dengan jenis data (numerik vs kategorikal).

6. **Eksplorasi Hubungan Antara Variabel**:
   - Identifikasi hubungan antara variabel-variabel dalam dataset menggunakan scatter plot, heatmap, atau pair plot.
   - Gunakan koefisien korelasi untuk mengukur hubungan linier antara variabel numerik.

7. **Penemuan Pola atau Insight**:
   - Identifikasi pola menarik, anomali, atau insight menarik dalam data.
   - Gunakan teknik seperti clustering atau analisis asosiasi untuk menemukan pola yang lebih kompleks.

8. **Validasi Asumsi atau Hipotesis**:
   - Jika ada, uji asumsi atau hipotesis yang mungkin berlaku dalam dataset.
   - Gunakan teknik statistik atau inferensi untuk menguji hipotesis.

9. **Iterasi dan Perbaikan**:
   - Iterasikan langkah-langkah EDA jika diperlukan untuk memperdalam pemahaman Anda tentang data.
   - Perbaiki dan perbaiki masalah yang Anda temui selama proses EDA.

10. **Dokumentasi Hasil**:
    - Dokumentasikan temuan dan insight penting dari analisis data.
    - Persiapkan laporan atau presentasi yang jelas dan informatif untuk berbagi hasil EDA dengan pemangku kepentingan lainnya.

ChatGPT Prompt: Beri kode untuk menampilkan pie diagram chart untuk fitur Age dengan min 20, max 69

```python
import matplotlib.pyplot as plt
import pandas as pd

# menentukan rentan nilai umur
min_age = 20
max_age = 69

# mengelompokkan data umur ke dalam kategori
bins = [min_age, 30, 40, 50, 60, max_age]
labels = ['20-30', '31-40', '41-50', '51-60', '61+']
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Menghitung jumlah individu dalam setiap kelompok usia
age_counts = df['Age_group'].value_counts()

# Membuat pie chart
plt.figure(figsize=(8, 6))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Usia')
plt.axis('equal')  # Memastikan lingkaran memiliki aspek rasio yang sama
plt.show()


```

Buat kode untuk menampilkan Histogram untuk setiap fitur : "Experience in Germany", "Seniority level",  "Yearly brutto salary", "Yearly bonus and stocks", "Previous yearly brutto salary", "Previous yearly bonus and stocks", "Vacation days"

```python
import matplotlib.pyplot as plt

# List fitur yang ingin diplot sebagai histogram
features = ["Experience in Germany", "Seniority level", "Yearly brutto salary",
            "Yearly bonus and stocks", "Previous yearly brutto salary",
            "Previous yearly bonus and stocks", "Vacation days"]

# Membuat subplot untuk setiap histogram
fig, axs = plt.subplots(len(features), 1, figsize=(10, 8*len(features)))

# Melakukan loop untuk setiap fitur dan membuat histogram
for i, feature in enumerate(features):
    axs[i].hist(df[feature].dropna(), bins=20, color='skyblue', edgecolor='black')
    axs[i].set_title(f'Histogram of {feature}')
    axs[i].set_xlabel(feature)
    axs[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Buat kode untuk menampilkan Box Plot untuk setiap fitur : "Experience in Germany", "Seniority level",  "Yearly brutto salary", "Yearly bonus and stocks", "Previous yearly brutto salary", "Previous yearly bonus and stocks", "Vacation days"

```python
import matplotlib.pyplot as plt

# List fitur yang ingin diplot sebagai box plot
features = ["Experience in Germany", "Seniority level", "Yearly brutto salary",
            "Yearly bonus and stocks", "Previous yearly brutto salary",
            "Previous yearly bonus and stocks", "Vacation days"]

# Membuat subplot untuk setiap box plot
fig, axs = plt.subplots(len(features), 1, figsize=(10, 8*len(features)))

# Melakukan loop untuk setiap fitur dan membuat box plot
for i, feature in enumerate(features):
    axs[i].boxplot(df[feature].dropna())
    axs[i].set_title(f'Box Plot of {feature}')
    axs[i].set_xlabel(feature)
    axs[i].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
