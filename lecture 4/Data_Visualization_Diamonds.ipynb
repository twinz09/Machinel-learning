```python
import pandas as pd

# mengunggah file cvs menjadi dataframe
df = pd.read_csv("/content/drive/MyDrive/Semester8/Machine Learning/lecture 4/dataset/diamonds.csv")

# menampilkan dataframe 5 baris pertama
print(df.head())
```

#Feature Statistic

```python
# menampilkan feature statistic
print("Feature Statistic:")
df.describe()
```

#Box Plot

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# menampilkan boxplot dari setiap fitur numerik
numerical_features= df.select_dtypes(include=['float64', 'int64']).columns
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.set_palette("pink")
    sns.boxplot(x=df[feature])
    plt.title('Boxplot of ' + feature)
    plt.show()
```

#Scatter Plot

```python
# memilih kolom
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# membuat scatter plots untuk setiap fitur numerik
sns.pairplot(df, vars=numeric_cols.columns, diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle('Scatter plot of Numerical Features', y=1.02, fontsize=16)
plt.show()
```

#Heat Map

```python
# memilih kolom numerik
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# menghitung matrik korelasi
correlation_matrix = numeric_cols.corr()

# menampilkan heat map
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Numerical Features', fontsize=16)
plt.show()
```
